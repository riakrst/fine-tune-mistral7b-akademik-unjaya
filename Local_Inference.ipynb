# Load model directly

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("riakrst/mistral-7b-pedoman-akademik-unjaya-merged")
model = AutoModelForCausalLM.from_pretrained("riakrst/mistral-7b-pedoman-akademik-unjaya-merged")
messages = [
    {"role": "user", "content": "apakah tugas dari seorang dosen pembimbing akademik?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
messages = [
    {"role": "user", "content": "Jelaskan secara ringkas tugas dosen pembimbing akademik sesuai dengan Pedoman Akademik Unjaya, sertakan sumber bab."},
]

inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs,
                        max_new_tokens=300,
                        temperature=0.2,
                        top_p=0.9)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

# Use a pipeline as a high-level helper

```python
from transformers import pipeline

pipe = pipeline("text-generation", model="riakrst/mistral-7b-pedoman-akademik-unjaya-merged")
messages = [
    {"role": "user", "content": "apakah tugas dari seorang dosen pembimbing akademik?"},
]
pipe(messages)
```

```python
messages = [
    {"role": "user", "content": "jelaskan tentang pedoman akademik"},
]
pipe(messages)
```

```python
messages = [
    {"role": "user", "content": "Jelaskan secara ringkas tugas dosen pembimbing akademik sesuai dengan Pedoman Akademik Unjaya, sertakan sumber bab."},
]
pipe(messages)
```

```python
messages = [
    {"role": "user", "content": "apakah tugas dari seorang dosen pembimbing akademik?"},
]
response = pipe(
    messages,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.1,
    do_sample=True
)

response
```

```python
messages = [
    {"role": "user", "content": "apakah tugas dari seorang dosen pembimbing akademik?"},
]

pipe(
    messages,
    max_new_tokens=300,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.1,
    do_sample=True
)
```
